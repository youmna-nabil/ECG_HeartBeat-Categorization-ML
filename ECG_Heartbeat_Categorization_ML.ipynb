```python
import pandas as pd

train = pd.read_csv('/content/mitbih_train.csv',header=None)
test = pd.read_csv('/content/mitbih_test.csv',header=None)

train.shape, test.shape
```

```python
train.head()
```

#Rename Columns

```python
train_df = pd.DataFrame(train)
test_df = pd.DataFrame(test)

del train, test

train_df.columns = [f'Feature {i + 1}' for i in range(len(train_df.columns) - 1)] + ['Target']
test_df.columns = [f'Feature {i + 1}' for i in range(len(test_df.columns) - 1)] + ['Target']
train_df
```

```python
# Create binary target variable for class 0
train_df['BinaryTarget'] = train_df['Target'].apply(lambda x: 0 if x == 0 else 1)

# Create multi-class target variable for classes 1, 2, 3, 4
train_df['multi_class_target'] = train_df['Target'].apply(lambda x: x if x in [1, 2, 3, 4] else None)
```

```python
train_df['BinaryTarget'].value_counts()
```

```python
train_df['multi_class_target'].value_counts()
```

```python
test_df['BinaryTarget'] = test_df['Target'].apply(lambda x: 0 if x == 0 else 1)
test_df['multi_class_target'] = test_df['Target'].apply(lambda x: x if x in [1, 2, 3, 4] else None)
```

```python
train_df['multi_class_target'].value_counts()
```

```python
test_df['BinaryTarget'].value_counts()
```

#Binary Classification

```python
X_train = train_df.drop(columns=['BinaryTarget', 'multi_class_target'])
y_train = train_df['BinaryTarget']
X_test = test_df.drop(columns=['BinaryTarget', 'multi_class_target'])
y_test = test_df['BinaryTarget']
```

```python
from sklearn.model_selection import train_test_split
X_binary, X_test_binary, y_binary, y_test_binary = train_test_split(X_train, y_train, test_size=0.2, random_state=42, stratify=y_train)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=0.95)),
    ('classifier', XGBClassifier())
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

print("Confusion Matrix (Binary Classifier):")
print(confusion_matrix(y_test, y_pred))
```

#MultiClass Classification

```python
train_df.dropna(subset=['multi_class_target'], inplace=True)
test_df.dropna(subset=['multi_class_target'], inplace=True)
```

```python
X_train = train_df.drop(columns=['BinaryTarget', 'multi_class_target'])
y_train = train_df['multi_class_target']
X_test = test_df.drop(columns=['BinaryTarget', 'multi_class_target'])
y_test = test_df['multi_class_target']
```

```python
from sklearn.model_selection import train_test_split
X_multi, X_test_multi, y_multi, y_test_multi = train_test_split(X_train, y_train, test_size=0.2, random_state=42, stratify=y_train)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
multi_Pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=0.95)),
    ('classifier', RandomForestClassifier())
])
```

```python
multi_Pipeline.fit(X_train, y_train)
```

```python
y_pred = multi_Pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

print("Confusion Matrix (MultiClass Classifier):")
print(confusion_matrix(y_test, y_pred))
```
